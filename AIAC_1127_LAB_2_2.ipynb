#Task 1: Cleaning Sensor Data

```python
def remove_negative_numbers(input_list):

    print(f"Original list: {input_list}")
    cleaned_list = [num for num in input_list if num >= 0]
    print(f"Cleaned list (without negative numbers): {cleaned_list}")
    return cleaned_list

# Sample usage:
sample_data = [1, -2, 3, 0, -5, 6, -10, 8]
cleaned_data = remove_negative_numbers(sample_data)

# Another sample:
sample_data_2 = [-10, -20, -30, 5, 0]
cleaned_data_2 = remove_negative_numbers(sample_data_2)
```

Task 2: String Character Analysis.

```python
def analyze_string(input_string):
    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0
    digit_count = 0

    for char in input_string:
        if char.isalpha():  # Check if the character is an alphabet
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
        elif char.isdigit():  # Check if the character is a digit
            digit_count += 1

    print(f"Analyzing string: '{input_string}'")
    print(f"Vowels: {vowel_count}")
    print(f"Consonants: {consonant_count}")
    print(f"Digits: {digit_count}")
    print("---------------------")
    return vowel_count, consonant_count, digit_count

# Sample usage:
analyze_string("Hello World 123")
analyze_string("Python Programming 4 fun")
analyze_string("NoVowelsHere 789")
analyze_string("aBcDeFg12")
```

### How the `analyze_string` function works:

1.  **Initialization**: It defines a string `vowels` containing all uppercase and lowercase vowels and initializes counters for `vowels`, `consonants`, and `digits` to zero.
2.  **Iteration**: It iterates through each character in the input string.
3.  **Character Classification**: For each character:
    *   It first checks if the character is an alphabet using `char.isalpha()`.
        *   If it's an alphabet, it then checks if it's present in the `vowels` string. If yes, `vowel_count` is incremented; otherwise, `consonant_count` is incremented.
    *   If the character is not an alphabet, it checks if it's a digit using `char.isdigit()`. If yes, `digit_count` is incremented.
4.  **Output**: Finally, it prints the counts of vowels, consonants, and digits found in the string.

Task 3: Palindrome Check.

```python
def is_palindrome(input_string):
    # Convert the string to lowercase and remove non-alphanumeric characters
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())

    # Compare the cleaned string with its reverse
    return cleaned_string == cleaned_string[::-1]

print(f"'madam' is a palindrome: {is_palindrome('madam')}")
print(f"'A man, a plan, a canal: Panama' is a palindrome: {is_palindrome('A man, a plan, a canal: Panama')}")
print(f"'hello' is a palindrome: {is_palindrome('hello')}")
print(f"'Racecar' is a palindrome: {is_palindrome('Racecar')}")
print(f"'12321' is a palindrome: {is_palindrome('12321')}")
```

### How the `is_palindrome` function works:

1.  **Clean the string**: It converts the input string to lowercase and removes any non-alphanumeric characters (like spaces, commas, etc.). This ensures that the comparison is case-insensitive and ignores punctuation.
2.  **Reverse and Compare**: It then compares this cleaned string with its reverse. If they are identical, the original string is a palindrome.
